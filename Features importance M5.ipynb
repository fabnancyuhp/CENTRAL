# Random Forest

```python
#Importation du moodel RDFjoblib
from joblib import dump, load
import numpy as np


rdf = load("C:/Users/IFDU1270/Desktop/MODELE520200814/RDFUNBA.joblib") 
```

```python
#on extrait les features importances:
features_importances = rdf.feature_importances_

#On extrait les indices par odre d√©croissant
importance = np.flip(np.argsort(features_importances))
feature_plus_important = features_importances[importance]


#On extrait le nom des colonnes
features_name = np.array(rdf.columns)
features_name_plus_iportant = features_name[importance]
```

### Les 20 features les plus importants

```python
features_name_20 = features_name_plus_iportant[0:20]
feature_plus_values_20 = feature_plus_important[0:20]
```

```python
list(zip(features_name_20.tolist(),(np.round(100*feature_plus_values_20,2)).tolist()))
```

```python
import matplotlib.pyplot as plt
y_pos = np.arange(len(features_name_20))

plt.bar(y_pos,np.round(feature_plus_values_20*100,2),edgecolor='navy',color='blue',width=0.9)
plt.xticks(y_pos,features_name_20,color='black',rotation=80)


plt.box(False)
plt.yticks([])
for i in range(0,len(features_name_20)):
    plt.text(y_pos[i]-0.25,np.round(feature_plus_values_20*100,1)[i]+0.25,str(np.round(feature_plus_values_20*100,1)[i]))
plt.show()
```
