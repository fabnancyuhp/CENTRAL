```python
import nltk
import re
import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

#nltk.download('punkt')
#nltk.download('stopwords')
```

```python
movie2['reviewbis'] = movie2['review'].apply(lambda x :x.lower())
movie2['reviewbis'] = movie2['reviewbis'].apply(lambda x : re.sub('\d+',' ',x))
movie2['reviewbis'] = movie2['reviewbis'].apply(lambda x : re.sub(r'[^\w]', ' ',x))
movie2['reviewbis'] = movie2['reviewbis'].apply(lambda x : re.sub(' +', ' ', x))
movie2['reviewbis'] = movie2['reviewbis'].apply(lambda x : re.sub(r'\S*@\S*\s?', ' ',x))
```

```python
#pour comprendre cette ligne
nlp_en = spacy.load("en_core_web_sm")
doc = nlp_en(movie2['review'][0])
doc
```

```python
#[token.lemma_ for token in doc]
import string
stop_plus_punk = list(string.punctuation)+stopwords.words('english')
[o for o in word_tokenize(movie2['review'][0]) if o not in stop_plus_punk]
```

```python
#On enlève les stopword. On peut utiliser word_tokenize
#nlp_en = spacy.load("en_core_web_sm")
import string
stop_plus_punk = list(string.punctuation)+stopwords.words('english')
def remov_stop(x):
    xbis = " ".join([o for o in word_tokenize(x) if o not in stop_plus_punk])
    return(xbis)

remov_stop(movie2['review'][0])
```

```python
#ici on applique la fonction remov_stop
movie2['reviewbis'] = movie2['reviewbis'].apply(lambda x : remov_stop(x))
```
